```python
df1=Airliner.copy()
# Converting the index as date
df1.index = pd.to_datetime(df1.index)
#df1.index = df1.index.date
#add year and month
df1['Year'] = pd.DatetimeIndex(df1.index).year
df1['Month'] = pd.DatetimeIndex(df1.index).month
df1['Day'] = pd.DatetimeIndex(df1.index).day
df1.tail(2)
```

```python
# delete last rwa 2022 for luck of data
#df1=df1.iloc[:-1,:]
```

```python
monthly_correlations_dict = {}
for month in df1.Month.unique():
    month_raw = df1[df1.index.month == month].iloc[:,:-3]
    monthly_corr = month_raw.corr()
    monthly_corr.reset_index(inplace=True)
    monthly_correlations_dict[month] = monthly_corr

monthly_correlations_df=pd.DataFrame.from_dict({(i,j): monthly_correlations_dict[i][j] 
                           for i in monthly_correlations_dict.keys() 
                           for j in monthly_correlations_dict[i].keys()},
                       orient='index')
monthly_correlations_df.columns= monthly_correlations_df.iloc[0,:]
monthly_correlations_df=monthly_correlations_df.iloc[1:,:]
df=monthly_correlations_df.fillna(0)
```

```python
df1.Year.unique()
```

```python
# Day BY MONTH

year_month_correlations_dict = {}
for year in df1.Year.unique(): 
    year_raw = df1[df1.index.year == year].iloc[:,:-3]
    for month in df1.Month.unique():

        year_month_raw = year_raw[year_raw.index.month == month]
        year_month_corr = year_month_raw.corr()
        year_month_corr.reset_index(inplace=True)
        year_month_correlations_dict[(year, month)] = year_month_corr

        year_month_correlations_df=pd.DataFrame.from_dict({(i,j): year_month_correlations_dict[i][j] 
                           for i in year_month_correlations_dict.keys() 
                           for j in year_month_correlations_dict[i].keys()},
                       orient='index')
year_month_correlations_df.columns= year_month_correlations_df.iloc[0,:]
year_month_correlations_df=year_month_correlations_df.reset_index()
year_month_correlations_df
```

```python
year_month_correlations_df['Year'], year_month_correlations_df['Month'] = year_month_correlations_df.level_0.str
```

```python
year_month_correlations_df
```

```python
yearly_corr = pd.DataFrame()

for y in  year_month_correlations_df.Year.unique():
    bloc_y=year_month_correlations_df[year_month_correlations_df['Year']==y]
    print(y,'-----------------------------------', bloc_y['level_0'].iloc[1])
    for m in bloc_y.Month.unique(): 
        bloc_y_m=bloc_y[bloc_y['Month']==m].iloc[:,1:-2]

        values = []
        column_names = []
        
        for i in range(1,len(bloc_y_m.columns)):
            for j in range(i-1,len(bloc_y_m.columns)):
                name1=bloc_y_m.iloc[j, 0] #raw , 0
                name2=bloc_y_m.columns[i] #col
                full_name = [f"{name1}_{name2}"]
                value =bloc_y_m.iloc[j, i] #raw , col        
                #print(i,j,name1, name2, value)
                values.append(value)
                column_names += full_name
        
        values.append(m)
        values.append(y)
        column_names += ['Month']
        column_names += ['Year']

        corr_part = pd.DataFrame(data=[values], columns=column_names)
        yearly_corr = pd.concat([yearly_corr, corr_part])
```
